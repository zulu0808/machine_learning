# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import re
import pandas as pd
!pip install python-dotenv
from dotenv import load_dotenv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from newsapi.newsapi_client import NewsApiClient

%matplotlib inline
```

```python
# Read your api key environment variable
# YOUR CODE HERE!
api_key = os.getenv('news_api')
```

```python
# Create a newsapi client
# YOUR CODE HERE!
newsapi = NewsApiClient (api_key = api_key)
```

```python
# Fetch the Bitcoin news articles
# YOUR CODE HERE!
path = ("https://newsapi.org/v2/everything?q=bitcoin&apiKey=6935a17225764822903bb074416d4e80")
btc_news = pd.read_json (path)
btc_news.head()
```

```python
# Fetch the Ethereum news articles
# YOUR CODE HERE!
path_2 = ("https://newsapi.org/v2/everything?q=ethereum&apiKey=6935a17225764822903bb074416d4e80")
eth_news = pd.read_json(path_2)
eth_news.head()
```

```python
# Create the Bitcoin sentiment scores DataFrame
# YOUR CODE HERE!
btc_sentiments = []

for article in btc_news["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        btc_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
        })

    except AttributeError:
        pass
```

```python
# Create Bitcoin DataFrame
btc_df = pd.DataFrame(btc_sentiments)
btc_df.head()
```

```python
# Reorder Bitcoin DataFrame columns
cols = ["compound", "negative", "neutral", "positive", "text"]
btc_df = btc_df[cols]
btc_df.head()
```

```python
# Create the ethereum sentiment scores DataFrame
# YOUR CODE HERE!
eth_sentiments = []

for article in eth_news["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        eth_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
        })

    except AttributeError:
        pass
```

```python
# Create Ethereum DataFrame
eth_df = pd.DataFrame(eth_sentiments)
eth_df.head()
```

```python
# Reorder Ethereum DataFrame columns
cols = ["compound", "negative", "neutral", "positive", "text"]
eth_df = eth_df[cols]
eth_df.head()
```
